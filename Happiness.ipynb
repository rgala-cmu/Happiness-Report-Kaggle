```python
df_2017 = pd.read_csv("../input/world-happiness/2017.csv")
df_2016 = pd.read_csv("../input/world-happiness/2016.csv")
df_2015 = pd.read_csv("../input/world-happiness/2015.csv")
df_2018 = pd.read_csv("../input/world-happiness/2018.csv")
df_2019 = pd.read_csv("../input/world-happiness/2019.csv")

print(df_2019.head())
df_2015 = df_2015.set_index('Country')
df_2016 = df_2016.set_index('Country')
df_2017 = df_2017.set_index('Country')
df_2018 = df_2018.set_index('Country or region')
df_2019 = df_2019.set_index('Country or region')

```

```python

df_merged = pd.concat([df_2015['Happiness Rank'],df_2016['Happiness Rank'],df_2017['Happiness.Rank'],df_2018['Overall rank'],df_2019['Overall rank']], axis=1, join='inner')
df_merged.head()
```

```python
df_merged['biggestChange'] = -df_2019['Overall rank'] + df_2015['Happiness Rank']

df_merged.sort_values('biggestChange')
```

```python
df_merged['biggestChange'].describe()
```

```python
sns.boxplot(df_merged['biggestChange'])
```
